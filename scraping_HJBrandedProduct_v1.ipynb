```python
import requests
from bs4 import BeautifulSoup
from bs4.element import Comment
import pandas as pd
```

```python
hj_branded1 = []
for page_number in range(1, 4):
    page_response = requests.get(f'http://www.hexpharmjaya.com/hj-branded?&page={page_number}')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    product_names = page_soup.select('.text-uppercase')
    for product_name in product_names:
        row = product_name
        hj_branded1.append(row)
    hj1=pd.DataFrame(hj_branded1, columns=["PRODUCT_NAME"]) 
```

```python
hj_branded2 = []
for page_number in range(1, 4):
    page_response = requests.get(f'http://www.hexpharmjaya.com/hj-branded?&page={page_number}')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    product_categorys = page_soup.select('.cat-desc p')
    for product_category in product_categorys:
        row = product_category
        hj_branded2.append(row)
    hj3=pd.DataFrame(hj_branded2, columns=["PRODUCT_CATEGORY"]) 
```

```python
frame = [ hj1,hj3]
result = pd.concat(frame,axis=1)
```

```python
result['PRODUCT'] = 'HJ BRANDED'
```
